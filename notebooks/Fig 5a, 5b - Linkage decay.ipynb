```r
files = Sys.glob("../data/all/linkage/*.tsv")

for (f in files){
    genome = strsplit(strsplit(f, split = '/')[[1]][5], split='\\.')[[1]][1]
    
    d = fread(f, sep="\t")
    d$r2 = d$r2_normalized
    d_n = d[which(d$mutation_type_A == 'N' & d$mutation_type_B == 'N'),]
    d_s = d[which(d$mutation_type_A == 'S' & d$mutation_type_B == 'S'),]
    d_ns = d[which((d$mutation_type_A == 'S' & d$mutation_type_B == 'N') |
                   d$mutation_type_A == 'N' & d$mutation_type_B == 'S'),]

    d_n$type = 'N'
    d_s$type = 'S'
    test = rbind(d_n, d_s)
    test$freq_a = (test$countab + test$countaB) / test$total
    test$freq_b = (test$countab + test$countAb) / test$total
    s = summary(glm(test$d_prime ~ test$type + test$distance + test$freq_a + test$freq_b))
    print(genome)
    print(s$coefficients)
    breaks <- seq(0,max(d$distance),25)

    n_by_mean <- aggregate(d_n$r2, FUN=mean,
                    by=list(distance=cut(d_n$distance, breaks, include.lowest=T)), na.rm=TRUE)
    n_by_length <- aggregate(d_n$r2, FUN=length,
                    by=list(distance=cut(d_n$distance, breaks, include.lowest=T)))
    linkage_n <- merge(n_by_mean, n_by_length, by = "distance")
    linkage_n = linkage_n[order(linkage_n$distance),]
    linkage_n$count = seq(1:nrow(linkage_n))

    linkage_n = linkage_n[which(linkage_n$x.y > 500),]
    linkage_n$type = 'N-N'

    s_by_mean <- aggregate(d_s$r2, FUN=mean,
                    by=list(distance=cut(d_s$distance, breaks, include.lowest=T)), na.rm=TRUE)
    s_by_length <- aggregate(d_s$r2, FUN=length,
                    by=list(distance=cut(d_s$distance, breaks, include.lowest=T)))
    linkage_s <- merge(s_by_mean, s_by_length, by = "distance")
    linkage_s = linkage_s[order(linkage_s$distance),]
    linkage_s$count = seq(1:nrow(linkage_s))

    linkage_s = linkage_s[which(linkage_s$x.y > 500),]
    linkage_s$type = 'S-S'

    ns_by_mean <- aggregate(d_ns$r2, FUN=mean,
                    by=list(distance=cut(d_ns$distance, breaks, include.lowest=T)), na.rm=TRUE)
    ns_by_length <- aggregate(d_ns$r2, FUN=length,
                    by=list(distance=cut(d_ns$distance, breaks, include.lowest=T)))
    linkage_ns <- merge(ns_by_mean, ns_by_length, by = "distance")
    linkage_ns = linkage_ns[order(linkage_ns$distance),]
    linkage_ns$count = seq(1:nrow(linkage_ns))

    linkage_ns = linkage_ns[which(linkage_ns$x.y > 500),]
    linkage_ns$type = 'N-S'


    linkage <- rbind(linkage_s, linkage_n)
    linkage <- rbind(linkage, linkage_ns)

    distances = list()
    for (dist in unlist(strsplit(as.character(linkage$distance), split=' '))){
        dist2 = gsub(strsplit(dist, split = ',')[[1]][2], pattern='\\]', replacement='')
        distances = c(distances, as.numeric(dist2))
    }
    linkage$Distance = as.numeric(distances)

    
    p <- ggplot(data=linkage, aes(x=Distance, y=x.x, group=type, colour=type, colour=type)) + 
    ggtitle(genome) + geom_point(aes(size=x.y), shape=22) + 
    scale_color_jcolors(palette = "pal5") + scale_fill_jcolors(palette = "pal5")+
    geom_smooth(alpha=0, aes(fill=type), span=0.97) + xlab("Genomic Distance (bp)") + ylab("r2")
    
#     pdf(file=paste(genome,'linkage.pdf', sep = ":"), width=8, height=4)
    print(p)
#     dev.off()
}
```

```r
### GAMETES 
gametes_n <- data.frame('genome' = character(), 'gamete2' = numeric(), 'gamete3' = numeric(),
                      'gamete4' = numeric(), stringsAsFactors = FALSE)

gametes_s <- data.frame('genome' = character(), 'gamete2' = numeric(), 'gamete3' = numeric(),
                      'gamete4' = numeric(), stringsAsFactors = FALSE)

for (f in files){
    genome = strsplit(strsplit(f, split = '/')[[1]][5], split='\\.')[[1]][1]
    
    d = fread(f, sep="\t")
    d$r2 = d$r2_normalized
    d_n = d[which(d$mutation_type_A == 'N' & d$mutation_type_B == 'N'),]
    d_s = d[which(d$mutation_type_A == 'S' & d$mutation_type_B == 'S'),]
    
    gametes = as.matrix(d_n[,c('countAB', 'countab', 'countAb', 'countaB')])
    gametes[which(gametes != 0)] <- 1
    gametes2 = as.numeric(table(rowSums(gametes))[2])
    gametes3 = as.numeric(table(rowSums(gametes))[3])
    gametes4 = as.numeric(table(rowSums(gametes))[4])

    gametes = as.matrix(d_s[,c('countAB', 'countab', 'countAb', 'countaB')])
    gametes[which(gametes != 0)] <- 1
    
    gametes2_s = as.numeric(table(rowSums(gametes))[2])
    gametes3_s = as.numeric(table(rowSums(gametes))[3])
    gametes4_s = as.numeric(table(rowSums(gametes))[4])
    
    gametes_n[nrow(gametes_n) + 1,] = list(genome, gametes2, gametes3, gametes4)
    gametes_s[nrow(gametes_s) + 1,] = list(genome, gametes2_s, gametes3_s, gametes4_s)
    
}
```

```r
names = read.csv('../data_tables/names.txt', sep="\t")
gametes_n = merge(gametes_n, names, by.x='genome', by.y='old')
gametes_s = merge(gametes_s, names, by.x='genome', by.y='old')

gametes_n[,c('gamete2', 'gamete3', 'gamete4')] = gametes_n[,c('gamete2', 'gamete3', 'gamete4')] / rowSums(gametes_n[,c('gamete2', 'gamete3', 'gamete4')])
gametes_s[,c('gamete2', 'gamete3', 'gamete4')] = gametes_s[,c('gamete2', 'gamete3', 'gamete4')] / rowSums(gametes_s[,c('gamete2', 'gamete3', 'gamete4')])
```
