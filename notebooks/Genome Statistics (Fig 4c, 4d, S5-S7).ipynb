```r
library(ggplot2)
library(jcolors)
setwd('/data6/Angelo/alexcc/AngeloStrainsPaper/manuscript_materials/soil_popgen/notebooks')
```

```r
d = read.csv('../data_tables/genome_summaries.txt')
d$snps_per_mb = (d$SNPs / d$genome_size) 
d$snps_per_mb = d$snps_per_mb * 1000000
```

```r
d
```

```r
d$freq_ratio = d$N_freq / d$S_freq
d$d_linkage_ratio = d$d_prime_n / d$d_prime_s
d$r_linkage_ratio = d$r2_n / d$r2_s
```

```r
names(d)
```

```r
?cor.test
```

```r
# pdf(file='corrmatrix.pdf')
library("corrplot")
data = d[,c('abundance','N.S', 'r2', 'd_prime', 'pi', 'genome_size', 'snps_per_mb', 'freq_ratio', 'd_linkage_ratio', 'r_linkage_ratio')]
res <- cor.mtest(data, conf.level = 0.99)
corrplot(cor(data), addCoef.col = "black", order = 'hclust', p.mat = res$p, insig = "blank", diag=FALSE)
# dev.off()
```

```r
# pdf(file='r2_vs_d.pdf',width=7,height=6)
ggplot(aes(x=d_prime,y=r2,size=pi), data=d) + 
geom_smooth(method='lm', alpha=0.1, color='black', show.legend = FALSE) + 
geom_point(aes(fill=phylum), shape=21)+ theme_bw() +
geom_text(aes(label=label, size=0.005),hjust=0.4, vjust=-1.3) + 
xlab("D'") + ylab("R2")+
scale_fill_manual(values = c('#33eba1', '#f0c033', '#fb6a6a', '#ab370d', '#9a653c', '#493c9a', '#cf6ec6', '#5e8dc9', '#0c882d'))
# dev.off()
```

```r
# pdf(file='d_vs_dns.pdf',width=7,height=6)
ggplot(aes(x=d_prime,y=d_linkage_ratio,size=pi), data=d) +
geom_smooth(method='lm', alpha=0.1, color='black', show.legend = FALSE) +
geom_point(aes(fill=phylum), shape=21)+ theme_bw() + 
geom_text(aes(label=label, size=0.005),hjust=0.4, vjust=-1.3) + 
xlab("D'") + ylab("N SNP D' / S SNP D'") + 
scale_fill_manual(values = c('#33eba1', '#f0c033', '#fb6a6a', '#ab370d', '#9a653c', '#493c9a', '#cf6ec6', '#5e8dc9', '#0c882d'))
# dev.off()
```
